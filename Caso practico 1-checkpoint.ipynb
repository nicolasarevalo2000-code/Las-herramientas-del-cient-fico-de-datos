```python
# ==========================================
# 1️⃣ IMPORTAR LIBRERÍAS
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")A


# ==========================================
# 2️⃣ CARGAR DATASETS
# (Usa la ruta correcta en tu computador)
# ==========================================

df_2015 = pd.read_csv(r"C:\Users\Nicolas\Desktop\Maestria\Las herramientas del científico de datos\Entregable\2015.csv")
df_2016 = pd.read_csv(r"C:\Users\Nicolas\Desktop\Maestria\Las herramientas del científico de datos\Entregable\2016.csv")

print("Primeras filas 2015:")
display(df_2015.head())

print("Primeras filas 2016:")
display(df_2016.head())


# ==========================================
# 3️⃣ COMPARAR COLUMNAS
# ==========================================

print("Columnas 2015:")
print(df_2015.columns)

print("\nColumnas 2016:")
print(df_2016.columns)


# ==========================================
# 4️⃣ UNIR DATAFRAMES
# ==========================================

df = pd.concat([df_2015, df_2016], axis=0, sort=False).reset_index(drop=True)

print("Dimensiones tras unir:", df.shape)


# ==========================================
# 5️⃣ ANÁLISIS DE NULOS
# ==========================================

nulos = df.isnull().sum()
porcentaje = (df.isnull().sum() / len(df)) * 100

resumen_nulos = pd.DataFrame({
    "Nulos": nulos,
    "Porcentaje (%)": porcentaje
})

display(resumen_nulos)


# ==========================================
# 6️⃣ REEMPLAZAR NULOS
# ==========================================

# Confidence Intervals → un único número aleatorio entre min y max
for col in ["Lower Confidence Interval", "Upper Confidence Interval"]:
    if col in df.columns:
        min_val = df[col].min()
        max_val = df[col].max()
        random_value = np.random.uniform(min_val, max_val)
        df[col] = df[col].fillna(random_value)

# Standard Error → media al cuadrado
if "Standard Error" in df.columns:
    mean_val = df["Standard Error"].mean()
    df["Standard Error"] = df["Standard Error"].fillna(mean_val**2)

# Eliminar cualquier otro nulo restante
df_clean = df.dropna()

print("Dimensiones después de limpiar nulos:", df_clean.shape)


# ==========================================
# 7️⃣ RESUMEN ESTADÍSTICO
# ==========================================

display(df_clean.describe())


# ==========================================
# 8️⃣ RELACIÓN FAMILIA VS SALUD
# ==========================================

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_clean, x="Family", y="Health (Life Expectancy)")
plt.title("Relación entre Familia y Salud")
plt.show()


# ==========================================
# 9️⃣ FELICIDAD VS CONFIANZA
# ==========================================

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_clean, x="Trust (Government Corruption)", y="Happiness Score")
plt.title("Felicidad vs Confianza en el Gobierno")
plt.show()


# ==========================================
# 🔟 MATRIZ DE CORRELACIÓN (CORREGIDO)
# SOLO COLUMNAS NUMÉRICAS
# ==========================================

plt.figure(figsize=(12,8))

corr = df_clean.select_dtypes(include=[np.number]).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Matriz de Correlación")
plt.show()


# ==========================================
# 1️⃣1️⃣ FELICIDAD MÁXIMA POR PAÍS
# ==========================================

df_max_happiness = df_clean.groupby("Country")["Happiness Score"].max().reset_index()

print("Felicidad máxima por país:")
display(df_max_happiness.head())


# ==========================================
# 1️⃣2️⃣ FELICIDAD VS GENEROSIDAD
# (Tamaño y color = Libertad)
# ==========================================

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df_clean,
    x="Generosity",
    y="Happiness Score",
    size="Freedom",
    hue="Freedom",
    palette="viridis"
)

plt.title("Felicidad vs Generosidad (Libertad como factor)")
plt.show()


# ==========================================
# 1️⃣3️⃣ DISTOPÍA POR REGIÓN
# ==========================================

plt.figure(figsize=(12,6))

sns.boxplot(data=df_clean, x="Region", y="Dystopia Residual")

plt.xticks(rotation=45)
plt.title("Distribución de Distopía por Región")
plt.show()
```
